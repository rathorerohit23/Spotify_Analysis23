```python
tracks.set_index("release_date", inplace = True)
tracks.index= pd.to_datetime(tracks.index)
tracks.head()
```

```python
tracks["duration"] = tracks["duration_ms"].apply(lambda x : round (x/1000))
tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
corr_tracks = tracks.drop(["key","explicit","mode"], axis=1)
```

```python
print(len(corr_tracks))
```

```python
corr_tracks = corr_tracks.copy()
```

```python
corr_tracks.head(10)
```

```python
sample_tracks = corr_tracks.sample(int(0.004*len(corr_tracks)))
```

```python
print(len(sample_tracks))
```

```python
sample_tracks.info()
```

```python
sample_tracks.head()
```

### Energy & Loudness Relation

```python
plt.figure(figsize=(12,8))
sns.regplot(data= sample_tracks, x= "energy", y ="loudness", scatter_kws={"color": "red"}, line_kws={"color": "black"}).set(title = "Energy & Loudness Co-relation")
```

### Popularity & Acousticness Relation

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sample_tracks, x= "acousticness", y ="popularity", scatter_kws={"color": "purple"}, line_kws={"color": "green"}).set(title = "Popularity & Acousticness Co-relation")
```

```python
sample_tracks["dates"] = sample_tracks.index.get_level_values('release_date')
sample_tracks.dates = pd.to_datetime(sample_tracks.dates)
years = sample_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title = "Number of Songs per Year")
```

```python
total_duration = sample_tracks.duration
fig_dim = (16,9)
fig, ax = plt.subplots(figsize =fig_dim)
fig = sns.barplot(x = years, y= total_duration, ax = ax, errwidth = False).set(title = "Years VS Duration")
plt.xticks(rotation=90)
```

```python
genre_data = pd.read_csv("D:/Tutorial Videos/Projects/Spotify/SpotifyFeatures.csv")
```

```python
genre_data.head()
```

```python
plt.title("Duration Of the Tracks by Genres")
sns.color_palette("Spectral", as_cmap= True)
sns.barplot(x= 'duration_ms', y = 'genre', data = genre_data)
plt.xlabel("Duration in ms")
plt.ylabel("Genres")
```

```python
sns.set_style=("grid")
plt.figure(figsize =(12,7))
fam_genre = genre_data.sort_values("popularity", ascending = False).head(10)
sns.barplot(y= 'genre', x = 'popularity', data = fam_genre).set(title = "Top Popular Genres ")
```

```python
top_five_artists = genre_data.groupby("artist_name").count().sort_values(by="track_name", ascending= False)["track_name"][:5]
top_five_artists
```

```python
top_ten_tracks = genre_data.groupby("track_name").count().sort_values(by="artist_name", ascending= False)["artist_name"][:10]
top_ten_tracks
```

```python
top_five_artists.plot.barh()
plt.show
```

```python
top_ten_tracks.plot.barh()
plt.show
```

```python
top_ten_dance_songs = genre_data[["danceability","track_name","artist_name"]].sort_values(by = "danceability", ascending = False)[:10]
top_ten_dance_songs
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x= "danceability", y = "artist_name", data= top_ten_dance_songs)
plt.title("ARTIST With MOST DANCEBLE SONGS")
plt.show()
```

```python
top_instrumental_tracks = genre_data[["instrumentalness","track_name","artist_name"]].sort_values(by = "instrumentalness", ascending = False)[:10]
top_instrumental_tracks
```

```python
plt.figure(figsize=(12,9))
plt.pie(x= "instrumentalness", data = top_instrumental_tracks, autopct= '%1.2f%%', labels= top_instrumental_tracks.track_name)
plt.title("Top 10 Instrumental Tracks")
plt.show()
```
